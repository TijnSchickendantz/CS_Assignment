```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
fi = pd.read_csv('istherecorrelation.csv', sep=';')
fi['Year']= pd.to_numeric(fi['Year'])
fi['NL Beer consumption [x1000 hectoliter]'] = pd.to_numeric(fi['NL Beer consumption [x1000 hectoliter]'])
```

```python
fig, ax = plt.subplots()
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()
p1, = ax.plot(fi['Year'], fi['WO [x1000]'], "b-", label="WO")
p2, = twin1.plot(fi['Year'], fi['NL Beer consumption [x1000 hectoliter]'], "r-", label="Beer consumption")

ax.set_xlabel("Year")
ax.set_ylabel("WO")
twin1.set_ylabel("Beer consumption in NL")
plt.show()
#print(np.corrcoef(fi['Year'], fi['WO [x1000]']))
```

```python
# We see some clear correlation between "WO" and "Beer consumption". Both variables increase over the years.
# So both variables "move" together. It could be very well be that this correlation is spurious, since it could 
# make sense that both variables just grow over the years.
```
